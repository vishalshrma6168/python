```python
from tkinter import *

def click(event):
    global scvalue
    Text=event.widget.cget("text") # widget--button btyega kon sa click hua hai,button se text dega
    if Text=="=":
        if scvalue.get().isdigit():
          value=int(scvalue.get())
           
        else:
            value=eval(scvalue.get())
            scvalue.set(value)
        screen.update()
        
    elif Text=="C":
        scvalue.set("")
        screen.update()   
    else:
        scvalue.set(scvalue.get() + Text)
        screen.update()


root=Tk()
root.title("Simple GUI")
root.geometry("644x600")

scvalue=StringVar()
scvalue.set("") 

# label
screen=Entry(root, textvar=scvalue, font="lucida 40 bold")
screen.pack(fill=X, pady=10, padx=10)

# frames for buttons
f1=Frame(root, bg="grey")

# buttons
b1=Button(f1,text="9",font="Lucidea 20 bold",padx=18,pady=20)
b1.pack(side=LEFT, padx=10, pady=10)
b1.bind("<Button-1>",click)

b1=Button(f1,text="8",font="Lucidea 20 bold",padx=18,pady=20)
b1.pack(side=LEFT, padx=10, pady=10)
b1.bind("<Button-1>",click)

b1=Button(f1,text="+",font="Lucidea 20 bold",padx=18,pady=20)
b1.pack(side=LEFT, padx=10, pady=10)
b1.bind("<Button-1>",click)

b1=Button(f1,text="=",font="Lucidea 20 bold",padx=18,pady=20)
b1.pack(side=LEFT, padx=10, pady=10)
b1.bind("<Button-1>",click)

b1=Button(f1,text="C",font="Lucidea 20 bold",padx=18,pady=20)
b1.pack(side=LEFT, padx=10, pady=10)
b1.bind("<Button-1>",click)

f1.pack()
root.mainloop()
```
