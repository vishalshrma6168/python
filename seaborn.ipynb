# bar plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# data
var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x=var.species, y=var.bill_length_mm)
plt.show()
```

# scatter plot

```python
# overlap 
sns.scatterplot(x=var.species, y=var.bill_length_mm)
plt.show()
```

# heatmap plot

```python
# no overlapping
sns.swarmplot(x=var.species, y=var.bill_length_mm)
plt.show()
```

```python
### point overlaps unitll jitter is applied
sns.stripplot(x=var.species, y=var.bill_length_mm,jitter=True)
plt.show()
```

```python
sns.countplot(x=var.species)
plt.show()
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.violinplot(x=var.species, y=var.bill_length_mm)
plt.show()
```
