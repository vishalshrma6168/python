#BAR_PLOT

```python
# bar plot
import matplotlib.pyplot as plt
import numpy as np
a=["a","B","C","D"]
b=[2,4,6,8]
x=[1,3,5]
y=[2,4,6]
c=["r"]
p=np.arange(len(a))
plt.xlabel("lang",fontsize=15)
plt.ylabel("num")
plt.title("graph")
plt.bar(p,b,width=0.3,color=c,align="center",edgecolor="r",linewidth=3,linestyle=":",alpha=0.5,label="home")
plt.legend()
# multiple bar graph
plt.bar(x,y,color=c)
plt.show()


# horizontal
# barh
```

# scatter plot

```python
# scatter plot
import matplotlib.pyplot as plt
a= [10,20,30,40]
b=[1,2,3,4]
size=[20,30,40,50]
plt.scatter(a,b,s=size,alpha=0.3,marker="*",edgecolors="g",linewidths=4)
# t= plt.colorbar()
# t.set_label("color")
plt.title("hello")
plt.xlabel("lang")
plt.ylabel("hello")
plt.show()
```

# multiple scatter plot in one diagram

```python
# multiple scatter plot in one diagram
import matplotlib.pyplot as plt
a= [10,20,30,40]
b=[1,2,3,4]
size=[20,30,40,50]
# 2nd garph
h=[3,7,7]
g=[5,8,9]
plt.scatter(a,b,s=size,alpha=0.7,marker="*",edgecolors="g",linewidths=4)
plt.scatter(h,g)
t= plt.colorbar()
t.set_label("color")
plt.title("hello")
plt.xlabel("lang")
plt.ylabel("hello")
plt.show()
```

# histogram plot

```python
# histogram plot
import matplotlib.pyplot as plt
import numpy as np
import random

x=np.random.randint(10,60,(10))
c=["g"]
plt.hist(x,color=c)
plt.title("hello")
plt.xlabel("num")
plt.ylabel("aaa")
plt.show()
```
