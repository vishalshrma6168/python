```python
from tkinter import *

def click(event):
  global scvalue
  Text=event.widget.cget("text")#text of the button clicked
  
  if Text=="=":
    if scvalue.get().isdigit():
      value=int(scvalue.get())
    else:
      value = eval(scvalue.get())#2*2=4
    scvalue.set(value)
    screen.update()
  elif Text=="C":
    scvalue.set("")
    screen.update()
  
  else:
    scvalue.set(scvalue.get() + Text)
    screen.update()

 
root = Tk()
root.title("Calculator")
root.geometry("644x900")
scvalue=StringVar()
scvalue.set("")

screen = Entry(root, textvar=scvalue, font="lucida 40 bold" )
screen.pack(fill=X, ipadx=8, pady=10, padx=10)

# frames for buttons
f1=Frame(root, bg="grey")

b=Button(f1, text="9", padx=20, pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=10, pady=10)
b.bind("<Button-1>",click)

b=Button(f1, text="8", padx=20,pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=10, pady=10)
b.bind("<Button-1>",click)

b=Button(f1, text="7", padx=20, pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=10, pady=10)
b.bind("<Button-1>",click)
f1.pack()

# frames for buttons
f1=Frame(root, bg="grey")
b=Button(f1, text="6", padx=20, pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=10, pady=10)
b.bind("<Button-1>",click)

b=Button(f1, text="5", padx=20, pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=10, pady=10)
b.bind("<Button-1>",click)

b=Button(f1, text="4", padx=20,pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=10, pady=10)
b.bind("<Button-1>",click)
f1.pack()
# frames for buttons
f1=Frame(root, bg="grey")

b=Button(f1, text="3", padx=20, pady=18, font="lucida 20 bold")

b.pack(side=LEFT, padx=10, pady=10)
b.bind("<Button-1>",click)
b=Button(f1, text="2", padx=20, pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=10, pady=10)
b.bind("<Button-1>",click)
b=Button(f1, text="1", padx=20,pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=10, pady=10)
b.bind("<Button-1>",click)
f1.pack()
# frames for buttons
f1=Frame(root, bg="grey")

b=Button(f1, text="0", padx=16, pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=19, pady=10)
b.bind("<Button-1>",click)
b=Button(f1, text="-", padx=20, pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=10, pady=10)
b.bind("<Button-1>",click)
b=Button(f1, text="*", padx=20, pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=10, pady=10)
b.bind("<Button-1>",click)
f1.pack()
# frames for buttons
f1=Frame(root, bg="grey")

b=Button(f1, text="/", padx=19, pady=18, font="lucida 20 bold")

b.pack(side=LEFT, padx=10, pady=10)
b.bind("<Button-1>",click)
b=Button(f1, text="%", padx=20, pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=10, pady=5)
b.bind("<Button-1>",click)
b=Button(f1, text="+", padx=20, pady=18, font="lucida 20 bold")
b.pack(side=LEFT, padx=10, pady=5)
b.bind("<Button-1>",click)
f1.pack()

# frames for buttons
f1=Frame(root, bg="grey")
b=Button(f1, text="C", padx=20, pady=30, font="lucida 20 bold")
b.pack(side=LEFT, padx=20, pady=5)
b.bind("<Button-1>",click)
b=Button(f1, text="=", padx=45, pady=30, font="lucida 20 bold")
b.pack(side=LEFT, padx=20, pady=5)
b.bind("<Button-1>",click)

f1.pack()
root.mainloop()
```
